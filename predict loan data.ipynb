# Setting up the Data

## Categorical Features

```python
purpose= pd.get_dummies(df['purpose'],drop_first=True)
purpose
```

```python
df1= pd.concat([df,pd.get_dummies(df['purpose'],drop_first=True)],axis=1)
df1.drop(['purpose'],axis=1,inplace=True)
df1
```

```python
pd.concat([df.drop(['purpose'],axis=1),pd.get_dummies(df['purpose'])],axis=1)
```

```python
pd.get_dummies(df,drop_first=True)
```

## Train Test Split


```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df1.drop(['not.fully.paid'],axis=1), df1['not.fully.paid'], test_size=0.20, random_state=101)
```

## Training a Decision Tree Model

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

## Predictions and Evaluation of Decision Tree

```python
pred = dtree.predict(X_test)
```

```python
df1_eval=pd.DataFrame({'Label': y_test,'Predicrions':pred})
df1_eval
```

```python
print(dtree.score(X_test,y_test))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,pred))
```

## Training the Random Forest model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=700)
```

```python
rfc.fit(X_train,y_train)
```

## Predictions and Evaluation of Random Forest

```python
rfc_pred = rfc.predict(X_test)
```

```python
df_eval=pd.DataFrame({'Label': y_test,'Predicrions':rfc_pred})
df_eval
```

```python
print(rfc.score(X_test,y_test))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,rfc_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,rfc_pred))
```
